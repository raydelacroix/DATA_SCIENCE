```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# Chargement du jeu de données Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# Réduction de dimension avec PCA (2 composantes)
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

# Réduction de dimension avec LDA (2 composantes)
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

# Affichage des résultats avant LDA
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
for color, i, target_name in zip(["navy", "turquoise", "darkorange"], [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, label=target_name)
plt.legend(loc="best")
plt.title("PCA du jeu de données Iris")
```

```python
# Affichage des résultats après LDA
plt.subplot(1, 2, 2)
for color, i, target_name in zip(["navy", "turquoise", "darkorange"], [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], color=color, alpha=0.8, label=target_name)
plt.legend(loc="best")
plt.title("LDA du jeu de données Iris")

plt.show()
```
