```python
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn import datasets
from sklearn.decomposition import PCA
from umap import UMAP

# Chargement du jeu de données Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# Réduction de dimension avec PCA (2 composantes)
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

# Réduction de dimension avec UMAP (2 composantes)
umap = UMAP(n_components=2)
X_umap = umap.fit_transform(X)
```

```python
# Affichage des résultats après UMAP
plt.figure(figsize=(12, 6))

# Affichage des résultats de PCA
plt.subplot(1, 2, 1)
for color, i, target_name in zip(["navy", "turquoise", "darkorange"], [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i,  1], color=color, alpha=0.8, label=target_name)
plt.legend(loc="best")
plt.title("PCA du jeu de données Iris")

# Affichage des résultats de UMAP
plt.subplot(1, 2, 2)
for color, i, target_name in zip(["navy", "turquoise", "darkorange"], [0, 1, 2], target_names):
    plt.scatter(X_umap[y == i, 0], X_umap[y == i,  1], color=color, alpha=0.8, label=target_name)
plt.legend(loc="best")
plt.title("UMAP du jeu de données Iris")

plt.show()
```
